You are a product manager and currently leading a team to develop a new product. Unfortunately, the latest version of your product fails the quality check. Since each version is developed based on the previous version, all the versions after a bad version are also bad.

Suppose you have n versions [1, 2, ..., n] and you want to find out the first bad one, which causes all the following ones to be bad.

You are given an API bool isBadVersion(version) which returns whether version is bad. Implement a function to find the first bad version. You should minimize the number of calls to the API.

 

Example 1:

Input: n = 5, bad = 4
Output: 4
Explanation:
call isBadVersion(3) -> false
call isBadVersion(5) -> true
call isBadVersion(4) -> true
Then 4 is the first bad version.

Example 2:

Input: n = 1, bad = 1
Output: 1

 

Constraints:

    1 <= bad <= n <= 231 - 1



```python
# The isBadVersion API is already defined for you.
# def isBadVersion(version: int) -> bool:
# Simulated API
def isBadVersion(version):
    return version >= 4  # First bad version is 4

    
class Solution:
    def firstBadVersion(self, n: int) -> int:
        """
        Find the first bad version in a list of versions.

        The list of versions is given as an integer n, where the range of versions is from 1 to n.

        The isBadVersion API is used to determine if a given version is bad or not.

        For example, if the list of versions is [1, 2, 3, 4, 5], and isBadVersion(3) returns True,
        and isBadVersion(2) returns False, then the first bad version is 3.

        The algorithm works by repeatedly dividing the search space in half and checking if the middle element is bad or not.
        If the middle element is bad, then the first bad version is in the left half, otherwise it is in the right half.
        The process is repeated until the search space is reduced to a single element, which is the first bad version.

        For example, if n = 5, then the search space is [1, 5]. The middle element is 3, which is bad, so the search space is reduced to [1, 3].
        The middle element of [1, 3] is 2, which is not bad, so the search space is reduced to [3, 3], which is the first bad version.

        Time complexity: O(log n)
        Space complexity: O(1)
        """
        left, right = 1, n # left and right are the boundaries of the search space
        while left < right: # while the search space is not empty
            mid = left + (right - left) // 2 # calculate the middle element of the search space
            if isBadVersion(mid): # if the middle element is bad
                right = mid # then the first bad version is in the left half
            else: # if the middle element is not bad
                left = mid + 1 # then the first bad version is in the right half
        return left # return the first bad version, which is the single element in the search space
```
